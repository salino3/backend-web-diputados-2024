```python
def get_new_tag(comunidad_tag, provincia_list, row_comunidades_tags_list, row_provincia_tags_list):
    if any(item in row_provincia_tags_list for item in provincia_list):
        # print(type(row_comunidades_tags_list))
        # print(type(comunidad_tag))

        # if comunidad_tag not in row_comunidades_tags_list:
        # print(row_comunidades_tags_list)
        out_row_comunidades_tags_list = row_comunidades_tags_list + [comunidad_tag]
       
        # else:
            # out_row_comunidades_tags_list = row_comunidades_tags_list            

            # print(row_comunidades_tags_list)
            # print(comunidad_tag)
        # print("-----------------------")
        # print(comunidad_tag)
        # print(provincia_list)
        # print(row_comunidades_tags_list)
        # print(row_provincia_tags_list)
        # print(out_row_comunidades_tags_list)
        
        # print('ok')
        # print("-----------------------")
    else:
        out_row_comunidades_tags_list = row_comunidades_tags_list
        # print("-----------------------")
        # print(comunidad_tag)
        # print(provincia_list)
        # print(row_comunidades_tags_list)
        # print(row_provincia_tags_list)
        # print(out_row_comunidades_tags_list)
        
        # print('No')       
        # print("-----------------------")
    return out_row_comunidades_tags_list
```

```python
def aplanar_lista(lista):
    resultado = []
    for elemento in lista:
        if isinstance(elemento, list):
            resultado.extend(aplanar_lista(elemento))  # Llamada recursiva
        else:
            resultado.append(elemento)
    return resultado
```
