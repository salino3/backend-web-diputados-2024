# Load libraries

```python
import requests
from bs4 import BeautifulSoup
# import pandas library
import pandas as pd
```

# Functions

```python
def has_sufficient_length(tag):
  """
  Aux. Function
  """
  return tag.name == 'meta' and len(tag.text) < 10
```

```python
def get_content(input):
  """
  Aux. Function
  """
  return input['content']
```

```python
def get_dict(code):
    """
    Main Function
    """
    #code
    code = str(code)
    #code
    new_code = code.zfill(6)
    #url
    url = "https://www.congreso.es/es/web/guest/busqueda-de-iniciativas?p_p_id=iniciativas&p_p_lifecycle=0&p_p_state=normal&p_p_mode=view&_iniciativas_mode=mostrarDetalle&_iniciativas_legislatura=XIV&_iniciativas_id=184/" + new_code
    # get_response
    response = requests.get(url, headers=headers)
    # Parsea el contenido HTML de la pÃ¡gina web
    soup = BeautifulSoup(response.text, 'html.parser')
    #...#
    long_paragraphs  = soup.find_all(has_sufficient_length)
    #...#
    long_paragraphs_filter  = long_paragraphs[2:5]
    #...#
    data = [get_content(i) for i  in long_paragraphs_filter]
    # dictionary with list object in values
    details = {
      'Bruto':data[0],
      'Tipo' : data[0].split(',')[0],
      'Presentada' : data[0].split(',')[1],
      'Calificada' : data[0].split(',')[2],
      'Contenido' : long_paragraphs_filter[2]['content'],
      'url':url
    }
    return details
```

```python
def get_data_dict(start,final):
    """
    Main Function
    """
    for i in range(start,final):
      print(f'{i}/120.000')
      try:
        new_dict[i] = get_dict(i)
        print('Existe')
      except:
        print('No existe')
        pass
    return new_dict
```

# Pipeline

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36',
}
new_dict = {}
```

```python
count = 0
for i in range(0, 120000, 100):
    count = count + 1
    start, final = (i, min(i + 100, 120000))
    data_dict = get_data_dict(start, final)
    df = pd.DataFrame(data_dict).T
    df.to_csv(f'congreso_preguntas_csv/out_{str(count).zfill(6)}.csv',sep=";, index=False)  
```
