# Load libraries

```python
import requests
from bs4 import BeautifulSoup
# import pandas library
import pandas as pd
import json
```

# Functions

```python
def get_diputado_dict(code, headers):
    """
    """
    #code
    code = str(code)
    #code
    new_code = code.zfill(3)
    #diputado_dict
    diputado_dict = {}
    #url
    url = f'''https://www.congreso.es/es/busqueda-de-diputados?p_p_id=diputadomodule&p_p_lifecycle=0&p_p_state=normal&p_p_mode=view&_diputadomodule_mostrarFicha=true&codParlamentario={new_code}&idLegislatura=XIV&mostrarAgenda=false'''
    # get_response
    response = requests.get(url, headers=headers)
    # Parsea el contenido HTML de la página web
    soup = BeautifulSoup(response.text, 'html.parser')
    # Get 'diputado_nombre_div'
    diputado_nombre_div = soup.find('div', class_='nombre-dip')
    # Obtener el texto dentro del div
    diputado_nombre_str = diputado_nombre_div.get_text(strip=True)  
    # Print the output 
    diputado_dict['nombre'] = diputado_nombre_str
    # Get 'diputado_nombre_div'
    diputado_grupo_parlamentario_div = soup.find('div', class_="grupo-dip")
    # Obtener el texto dentro del div
    diputado_grupo_parlamentario_str = diputado_grupo_parlamentario_div.get_text(strip=True)  
    # Print the output 
    diputado_dict['grupo_parlamentario'] = diputado_grupo_parlamentario_str
    diputado_dict['grupo_parlamentario'] = diputado_grupo_parlamentario_str.split('\r\n')[0]
    return diputado_dict
```

# Pipeline

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36',
}
```

```python
diputados_dictdict = {}
for i in range(1,450):
    print(i)
    try:
        diputados_dictdict[str(i).zfill(3)] = get_diputado_dict(str(i), headers)
    except:
        pass
```
